### Column Transformations <br>
<p>Rarely, if ever, will the columns in the original raw dataframe read from CSV or database table be the ones you actually need for your analysis. You will spend lots of time constantly transforming columns or groups of columns using general computational operations to produce new ones that are functions of the old ones. Pandas has full support for this: Consider the following dataframe containing membership term and renewal number for a group of customers:</p>

```python
#lecture @ 23 min mark
# Generate some fake data
customer_id = np.random.randint(1000,1100, 10)
renewal_nbr = np.random.randint(0,10,10)
customer_dict = {1: 0.5, 0: 1}
terms_in_years = [customer_dict[key] for key in np.random.randint(0,2,10)]

random_data = {
    'customer_id': customer_id,
    'renewal_nbr': renewal_nbr,
    'term_in_years': terms_in_years
}

customers = pd.DataFrame.from_dict(random_data)
customers


# later to increase sample size to demonstrate deleting duplicates:
#     customer_id = np.random.randint(1000,1500, 1000)
#     renewal_nbr = np.random.randint(0,10,1000)
```

##### User Defined Function

<p>If what you want to do to a column that can't be represented by simple mathematical operations, you can write your own $\textit{user defined function}$ with the full customizability available in Python and any external Python packages, then map it directly onto a column. Let's add some ages to our customer dataframe, and then classify them into our custom defined grouping scheme:</p>

```python
# use .apply to map over dataframe

# Create a new column for ages
customers['age'] = np.random.randint(18,80,10)
customers

# Create a new column 'age group' which will have values teenager, adult, and senior
    # use a udf - user defined function
    # when you're working with a UDF
        # either working with function accepting a SINGLE VALUE from a column
        # or a SINGLE ROW from dataframe
        # aka the input to UDF won't be the entire datafram - it will just be a single value or single row from dataframe
    # udf will be applied to each row separately

def make_age_groups(age):
    if age < 20:
        return 'Teenager'
    elif age < 35:
        return 'Young Adult'
    elif age < 65:
        return 'Adult'
    else:
        return 'Senior'
    
# is UDF acting on a single column? if so, .apply() it to that column
# is UDF acting on a row(aka multiple columns impact the result)? if so, .apply() it to the entire dataframe
    # example of this later
    
# apply this UDF in the creation of the 'age_group' column
customers['age_group'] = customers['age'].apply(make_age_groups)
# running through entire dataframe, and for every row, applying the value in age to the make-age-groups function to produce value for age_group
customers
```

<p>As a last example I'll show here how you would use a lambda function to create a UDF that depends on $\textit{more than one}$ column:</p>

<li>UDF = User Defined Function</li>

```python
# when we .apply() this UDF, we must specify the axis of application
# Axis for apply can only be 1 or 0 -- 1 being the X axis, 0 being the Y axis

# create a 'loyalty_status' column
# depending on tenure and age, customer willl be considered loyal or new

# normal UDF here:
def make_loyalty(row):
    # loyalty_coefficient will be age*tenure
    # customer is considered loyal if loyalty_coefficient is > 150
    if row['age']* row['tenure'] > 150:
        return 'loyal'
    return 'new'

# lambda version of same udf
mloyal = lambda row: 'loyal' if row['age']*row['tenure']> 150 else 'new'

# using the lambda version we can create this new column in a single line
# question - is this UDF dependent on a single value or more than one value?
    # answer - more than 1
        # therefore, we .apply() this UDF to the entire dataframe
# when you .apply() a UDF to an entire dataframe you must specify does this UDF act on each row or each column
    # by setting the 'axis' parameter to 1 for rows or 9 for columns
customers['loyalty_status'] = customers.apply( lambda row: 'loyal' if row['age']*row['tenure']> 150 else 'new', axis=1)
# setting axis =1 parameter means it goes on a row-by-row basis

# print entire table with new loyalty_status column
customers

# look at customers who qualify as 'loyal'
customers[customers['loyalty_status'] == 'loyal']
```

### Aggregations <br>
<p>The raw data plus some transformations is generally only half the story. Your objective is to extract actual insights and actionable conclusions from the data, and that means reducing it from potentially billions of rows to some summary statistics via aggregation functions.</p>

##### groupby() <br>
<p>The .groupby() function is in some ways a 'master' aggregation.</p> 

<p>Data tables will usually reserve one column as a primary key - that is, a column for which each row has a unique value. This is to facilitate access to the exact rows of a data table that a user wants to view. The other columns will often have repeated values, such as the age groups in the above examples. We can use these columns to explore the data using the Pandas API:</p>

```python
# also introducing .count() here, exact same as to how it's used in SQL

# Using the groupby with the column intact as a column/key
# customers.groupby('age_group', as_index = False).count()[['customer_id','age_group']]

customers.groupby('age_group')
    # returns: <pandas.core.groupby.generic.DataFrameGroupBy object at 0x000002C07AFC34C0>
    # .groupby() then needs an aggregate function to tell it how to group together the rows of data
    
customers.groupby('age_group').count()
    # returns number of people in each age group for each category
    
    
customers.groupby('age_group').count()['customer_id']
    # returns number of people in each age group by their customer_id
```

##### Type of groupby()

<p>The result is a new dataframe, the columns of which all contain the counts of the grouped field. Notice the type of a grouped dataframe:</p>

```python
# unless you have a paired aggregate function, grouping by doesnt make any sense
print(type(customers), 'Regular Data Frame')
print(f'\n Type of Group By for Customer.GroupBy: {type(customers.groupby("age_group"))}')
```

<p>This is because simply grouping data doesn't quite make sense without an aggregation function like count() to pair with. In this case, we're counting occurances of the grouped field, but that's not all we can do. We can take averages, standard deviations, mins, maxes and much more! Let's see how this works a bit more:</p>

##### mean()

```python
# mean = average
customers.groupby('age_group').mean()[['tenure','age']]
```

##### groupby() w/Multiple Columns

<p>We end up with the average age of the groups in the last column, the average tenure in the tenure column, and so on and so forth. You can even split the groups more finely by passing a list of columns to group by:</p>

```python
customers.groupby(['age_group', 'age']).count()[['customer_id']]

# first grouped by age_group, then counted by age within that 
# shows there are currently 2 68-yr-olds and 2 78-yr-olds within the senior group
```

```python
allstars # checking that it does exist
allstars.groupby('AllStar').mean()[['BA','OBP']]
```

##### drop_duplicates()

<p>Drops all duplicates from the current dataframe</p>
<li> will delete the second occurrence/row of that value(customer_id)

```python
customers.drop_duplicates('customer_id', inplace=True) # won't return anything - it's any inplace operation, trying to print it will just say none
# if its in place, you cant chain any other functionalities to end/back of it
customers.reset_index(inplace=True, drop=True) # resetting the index values, 
customers
```

```python
# if you accidentally rerun reset_index and end up with extra columns... how you remove a column?
# same way you remove a key from a dict
# del table[<column name>] ->> del customers['index'] to delete extra index columns
```

```python
# check group by again now that you've deleted any duplicate occurrences
customers.groupby(['age_group', 'age']).count()[['customer_id']]

# 5 ages with the most customers at that age
customers.groupby(['age_group', 'age']).count()[['customer_id']].sort_values('customer_id', ascending=False).head(5)
```

In one line above, we actually did something very complicated - that's the power of the dataframe.
